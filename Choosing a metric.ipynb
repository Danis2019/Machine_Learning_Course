# Библиотеки

```python
from sklearn import datasets # для boston
from sklearn import preprocessing # Для масштабирования признаков
from sklearn.neighbors import KNeighborsRegressor # kNN
from sklearn.neighbors import KNeighborsClassifier # kNN
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

# Загружаем выборку Boston
## B['data'] - признаки, B['target'] - ответы(цены)

```python
B = datasets.load_boston()
```

```python
print(B['data'])
print(B['target'])
```

# Масштабируем признаки

```python
X = preprocessing.scale(B['data'])
```

```python
y = B['target']
```

```python
P = np.linspace(1, 2, num=200)
```

```python
X
```

## Добавим кросс-валидацию

```python
kf = KFold(shuffle=True,random_state=42)
```

# Обучение

```python
a = []
for i in P:
    neigh = KNeighborsRegressor(n_neighbors=5, weights='distance', p = i, metric='minkowski')
    neigh.fit(X, y)
    a.append([i,np.mean(cross_val_score(neigh, X, y, scoring='neg_mean_squared_error'))])
```
