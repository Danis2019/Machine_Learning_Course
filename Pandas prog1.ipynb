## Какого возраста были пассажиры? Посчитайте среднее и медиану возраста пассажиров.

```python
age = np.array(data['Age'].dropna())# dropna() - удаление пустых значений
```

```python
mean = np.mean(age)
print(mean)
median = np.median(age)
print(median)
```

## Какое самое популярное женское имя на корабле?

```python
woman_names = []
for i in range(1, len(data['Name']) + 1):
    words = nltk.word_tokenize(data['Name'][i])
    for i in range(1,len(words)):
        if words[i] == 'Miss.':
            if words[i + 1] == '(':
                woman_names.append(words[i + 2])
            else:
                woman_names.append(words[i + 1])
        if words[i] == 'Mrs.':
            skobka = False
            for j in range(1,len(words)):
                if words[j - 1] == '(':
                    skobka = True
                    woman_names.append(words[j])
            if skobka == False:
                woman_names.append(words[i + 1])
print(woman_names)
```

```python
frame_woman_names = pd.DataFrame(woman_names)
```
